````
import pandas as pd
data = {
       '이름':['홍길동','이순신','강감찬'],
       '국어':[90,70,80],
       '영어':[99,77,88],
       '수학':[75,85,90]
       }
df = pd.DataFrame(data)
df = df.set_index('이름')
df
````

````
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] =False
df.plot()
plt.show()
````

## (4)막대 그래프 그리기

````
df.plot.bar(cmap='magma')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.title('막대 그래프',fontdict={'fontsize': 20, 'fontweight': 'bold', 'color': 'blue'})
plt.xlabel('이름')
plt.ylabel('점수')
plt.show()
````

## (5)원 그래프

````
import pandas as pd
df
````

````
#df['국어'].plot.pie(autopct='%.1f%%',startangle=90,colors=['pink','lightgreen','lightblue'])
df['국어'].plot.pie(autopct='%.1f%%',startangle=90,cmap='spring')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.title('국어 성적 그래프',size=20)
plt.ylabel('')
plt.show()
````

## (6)과목별 원그래프 그리기

````
fig,axes = plt.subplots(1,3,figsize=(18,6))

df['국어'].plot.pie(autopct='%.1f%%',startangle=90, ax=axes[0],cmap='spring')
df['영어'].plot.pie(autopct='%.1f%%',startangle=90, ax=axes[1],cmap='summer')
df['수학'].plot.pie(autopct='%.1f%%',startangle=90, ax=axes[2],cmap='winter')

plt.suptitle('과목별 성적 비율',size=20)
axes[0].set_title('국어 점수 비율',size=15)
axes[0].set_ylabel('')
axes[1].set_title('영어 점수 비율',size=15)
axes[1].set_ylabel('')
axes[2].set_title('수학 점수 비율',size=15)
axes[2].set_ylabel('')

plt.show()
````

## (7)과목별 막대 그래프 그리기

````
fig,axes = plt.subplots(1,3,figsize=(18,6))

df['국어'].plot.bar(ax=axes[0],color='red',alpha=0.5)
df['영어'].plot.bar(ax=axes[1],color='yellow',alpha=0.7)
df['수학'].plot.bar(ax=axes[2])

plt.suptitle('과목별 성적 그래프',size=20)

axes[0].set_title('국어 점수',size=15)
axes[0].set_xticklabels(df.index,rotation=0)
axes[0].set_facecolor('lightyellow')
axes[0].grid(linestyle='--',linewidth=0.5)
axes[0].legend(loc='upper center')

axes[1].set_title('영어 점수',size=15)
axes[1].set_xticklabels(df.index,rotation=0)
axes[1].set_facecolor('lightblue')
axes[1].grid(linestyle='--',linewidth=0.5)
axes[1].legend(loc='upper center')

axes[2].set_title('수학 점수',size=15)
axes[2].set_xticklabels(df.index,rotation=0)

plt.show()
````

##(12)종합 성적 막대 그래프 그리기

````
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] =False

df = df3
````

````
fig,ax=plt.subplots(figsize=(16,6))
plt.suptitle('종합성적', fontsize=16)
bars = df.plot.bar(ax=ax,cmap='spring')
plt.xticks(rotation=0)
plt.xlabel('이름')
plt.ylabel('점수')

for bar in bars.patches:
  y=bar.get_height()
  x=bar.get_x()+bar.get_width()/2
  ax.text(x,y+1,y,ha='center',fontsize=8)
plt.show()
````
